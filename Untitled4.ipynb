```python
import os
import xml.etree.ElementTree as ET
import numpy as np
from tqdm import tqdm
from PIL import Image
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# Define the paths to the dataset
dataDir = '/content/dataset/VOCdevkit/VOC2012'
imgDir = os.path.join(dataDir, 'JPEGImages')
annDir = os.path.join(dataDir, 'Annotations')

# PASCAL VOC classes
class_names = ['aeroplane', 'bicycle', 'bird', 'boat', 'bottle',
               'bus', 'car', 'cat', 'chair', 'cow', 'diningtable',
               'dog', 'horse', 'motorbike', 'person', 'pottedplant',
               'sheep', 'sofa', 'train', 'tvmonitor']

class_to_index = {cls: idx for idx, cls in enumerate(class_names)}

# Function to load images
def load_image(img_path):
    img = Image.open(img_path).convert('RGB')
    img = img.resize((128, 128))  # Resize to 128x128 for the model
    return np.array(img) / 255.0

# Function to load annotations
def load_annotations(ann_path):
    tree = ET.parse(ann_path)
    root = tree.getroot()
    labels = np.zeros(len(class_names))
    for obj in root.findall('object'):
        cls = obj.find('name').text
        if cls in class_to_index:
            labels[class_to_index[cls]] = 1
    return labels

# Load images and labels
X = []
y = []

img_files = os.listdir(imgDir)
ann_files = os.listdir(annDir)

for img_file in tqdm(img_files[:5000]):  # Limiting to 5000 images for the example
    img_path = os.path.join(imgDir, img_file)
    ann_path = os.path.join(annDir, img_file.replace('.jpg', '.xml'))
    if os.path.exists(ann_path):
        img_data = load_image(img_path)
        labels = load_annotations(ann_path)
        X.append(img_data)
        y.append(labels)

X = np.array(X)
y = np.array(y)

print(f'Loaded {len(X)} images')
```

```python
# Define the CNN model
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(len(class_names), activation='sigmoid'))

# Display the model's architecture
model.summary()

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X, y, epochs=10, validation_split=0.2)

# Plot the training history
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()

# Evaluate the model
test_loss, test_acc = model.evaluate(X, y, verbose=2)
print(f"Test Accuracy: {test_acc}")
```
